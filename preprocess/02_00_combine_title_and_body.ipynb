```python
import pandas as pd
data = pd.read_csv("../data/04_01_dummy_validation.csv")
data.head()
```

```python
data['text'] = data['title'] + data['body']
```

```python
data.head()
```

```python
data.text = data.text.replace('\s+', ' ', regex=True)
```

```python
# reorder the columns
cols = data.columns.tolist()
```

```python
type(cols)
```

```python
cols = cols[0:1] + cols[-1:] + cols[1:-1]
```

```python
cols
```

```python
data = data[cols]
data.head()
```

```python
drop_data = data.drop(['title', 'body'], axis=1)
```

```python
drop_data.tail()
```

```python
drop_data.to_csv("../data/04_02_processed_validation.csv", index=False)
```
