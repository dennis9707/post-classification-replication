```python
# 
all_data = []
for i in non_dup_list:
    tmp = result.loc[result['id'] == i]
    label = []
    change = 0
    usage = 0
    concep = 0	
    discre = 0 
    doc = 0
    errors = 0 
    review = 0
    # API_CHANGE	API_USAGE	CONCEPTUAL	DISCREPANCY	DOCUMENTATION	ERRORS	REVIEW
    for index, row in tmp.iterrows():
        qid = row['id']
        title = row['title']
        body = row['body']
        if row['API_CHANGE'] == 1:
            change = 1
        elif row['API_USAGE'] == 1:
            usage = 1
        elif row['CONCEPTUAL'] == 1:
            concep = 1
        elif row['DISCREPANCY'] == 1:
            discre = 1
        elif row['DOCUMENTATION'] == 1:
            doc = 1
        elif row['ERRORS'] == 1:
            errors = 1 
        elif row['REVIEW'] == 1:
            review = 1 
    single_data = [qid, title, body, change, usage, concep, discre, doc, errors, review]
    all_data.append(single_data)


new_df = pd.DataFrame(all_data, columns=['id', 'title', 'body', "API_CHANGE", "API_USAGE", "CONCEPTUAL", "DISCREPANCY", "DOCUMENTATION", "ERRORS", "REVIEW"])
```
